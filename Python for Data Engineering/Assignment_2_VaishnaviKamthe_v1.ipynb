```python
import pymysql

conn = pymysql.connect(host="localhost", user="root", password="root")
cursor = conn.cursor()
```

```python
# 1. Create a database.
cursor.execute("CREATE DATABASE IF NOT EXISTS mydb")
cursor.execute("USE mydb")
```

```python
# 2. Create a table with 3-4 columns.
# 3.      Add a primary key.
cursor.execute("""
CREATE TABLE IF NOT EXISTS student1 (
    id INT PRIMARY KEY,
    name VARCHAR(50),
    age INT,
    city VARCHAR(50)
)
""")
```

```python
# 4.      Insert 5 rows.
sql = "INSERT INTO student1 VALUES (%s, %s, %s, %s)"
data = [
    (1, "Amit", 22, "Pune"),
    (2, "Neha", 23, "Mumbai"),
    (3, "Raj", 21, "Delhi"),
    (4, "Sara", 24, "Bangalore")
]

cursor.executemany(sql, data)
conn.commit()
```

```python
# 5.      Update a row.
cursor.execute("UPDATE student1 SET age = 30 WHERE id = 1")
conn.commit()
```

```python
# 6.      Print all the rows of the table.
cursor.execute("SELECT * FROM student1")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
# 7.      Delete first 2 rows of the table.
cursor.execute("DELETE FROM student1 WHERE id IN (1,2)")
conn.commit()
```

```python
# 8.      Create another table with same data structure “create table as” query.
cursor.execute("CREATE TABLE copy_student LIKE student1")
```

```python
# 9.      Display columns and data types of the created table.
cursor.execute("DESCRIBE copy_student")
for col in cursor.fetchall():
    print(col)
```

```python
conn.close()
```

```python
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String

# 1.      Create a database.
engine = create_engine("mysql+pymysql://root:root@localhost/mydb")
conn = engine.connect()
metadata = MetaData()
```

```python
# 2.      Create a table with 3-4 columns. 
# 3.      Add a primary key.
student = Table('student_sa', metadata,
    Column('id', Integer, primary_key=True),
    Column('name', String(50)),
    Column('age', Integer),
    Column('city', String(50))
)

metadata.create_all(engine)
```

```python
# 4.      Insert 5 rows.
conn.execute(student.insert(), [
    {'id':1, 'name':'Amit', 'age':22, 'city':'Pune'},
    {'id':2, 'name':'Neha', 'age':23, 'city':'Mumbai'},
    {'id':3, 'name':'Raj', 'age':21, 'city':'Delhi'},
    {'id':4, 'name':'Sara', 'age':24, 'city':'Bangalore'},
    {'id':5, 'name':'John', 'age':25, 'city':'Chennai'}
])
```

```python
# 5.      Update a row.
conn.execute(student.update().values(age=35).where(student.c.id == 1))
```

```python
# 6.      Print all the rows of the table.
result = conn.execute(student.select()).fetchall()

for row in result:
    print(row)
```

```python
# 7.      Delete first 2 rows of the table.
conn.execute(student.delete().where(student.c.id.in_([1,2])))
```

```python
# 8.      Create another table with same data structure “create table as” query.
from sqlalchemy import text

conn.execute(text("CREATE TABLE student_sa_copy LIKE student_sa"))
```

```python
# 9.      Display columns and data types of the created table.
result = conn.execute(text("DESCRIBE student_sa_copy"))

for row in result:
    print(row)
```
