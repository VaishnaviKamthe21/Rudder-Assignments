```python
cars["Price"] = cars["Price"].replace('[\$,]', '', regex=True)
cars["Price"] = cars["Price"].astype(float)
print(cars["Price"])
```

```python
print("Max Price:", cars["Price"].max())

print("Min Price:", cars["Price"].min())

print("Sum:", cars["Price"].sum())

print("Mean:", cars["Price"].mean())

print("Median:", cars["Price"].median())
```

```python
cars.to_csv("car-sales-new.csv", index=False)
```

```python
engine = create_engine(
    "mysql+pymysql://root:root@localhost/car_db"
)

cars.to_sql(
    name="car_sales",
    con=engine,
    if_exists="replace",
    index=False
)

print("Data inserted into MySQL")
```

```python
filtered = cars[
    (cars["Price"] >= 10000) &
    (cars["Price"] <= 20000)
]

print(filtered)
```

```python
makes = cars["Make"].unique()

for make in makes:
    make_df = cars[cars["Make"] == make]
    filename = f"{make}.csv"
    make_df.to_csv(filename, index=False)
    print(filename, "created")
```

```python
data = pd.read_csv("data.csv")

data.head()
```

```python
n = int(input("Enter number of rows per file: "))
```

```python
rows = len(data)

file_number = 1

for start in range(0, rows, n):
    end = start + n
    split_data = data.iloc[start:end]
    filename = f"data_{file_number}.csv"
    split_data.to_csv(filename, index=False)
    print(filename, "created")
    file_number += 1
```
