# Churn Proportion

Approximately 56.7% of customers have churned, while 43.3% remain active. This indicates a relatively high churn rate, suggesting potential issues with retention or customer experience.

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation = df[num_cols].corr()['Churn'].sort_values(ascending=False)
correlation
```

# Key Features Correlated with Churn

**Support Calls** show the strongest positive correlation with churn, indicating that customers requiring frequent support are more likely to leave.

**Contract Length** has a significant impact, with customers on monthly contracts showing extremely high churn compared to longer-term contracts.

**Total Spend** is negatively correlated with churn, suggesting that higher-value customers are less likely to leave.

# Churn Intervention Strategy

Customers with high support interaction and those on monthly contracts show the highest likelihood of churn. A targeted intervention would be to proactively identify such customers and offer improved support experiences along with incentives to shift to longer-term contracts.

For example, customers with frequent support calls can be provided priority support or issue resolution follow-ups, while monthly users can be encouraged to switch to annual plans through discounts or added benefits.
