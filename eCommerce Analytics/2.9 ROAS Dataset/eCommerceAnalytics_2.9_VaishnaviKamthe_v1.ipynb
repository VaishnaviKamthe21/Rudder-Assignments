```python
import pandas as pd

df = pd.read_csv("data.csv")
df.head()
```

```python
df.info()
```

```python
df = df[df['approved_conversion'].notna()]

df['ROAS'] = df['approved_conversion'] / df['spent']
```

```python
df[['campaign_id', 'spent', 'approved_conversion', 'ROAS']].head()
```

# ROAS Calculation

Return on Ad Spend (ROAS) was calculated by dividing approved conversions by total ad spend. Since direct revenue data was not available, approved conversions were used as a proxy for revenue.

```python
campaign_roas = df.groupby('campaign_id').agg({
    'spent': 'sum',
    'approved_conversion': 'sum'
}).reset_index()

campaign_roas['ROAS'] = campaign_roas['approved_conversion'] / campaign_roas['spent']

campaign_roas
```

```python
campaign_roas = campaign_roas.sort_values(by='ROAS', ascending=False)
campaign_roas
```

# Campaign Ranking

Campaigns were ranked based on ROAS to identify the most and least efficient marketing investments.

# Campaign Performance Analysis

Campaign 916 shows the highest ROAS despite having the lowest spend, indicating efficient use of budget. Campaign 936 has moderate performance, while Campaign 1178 has the lowest ROAS despite the highest spend, suggesting inefficient spending.

Scale Down – Campaign 1178:

Despite high spending, this campaign delivers the lowest ROAS. This indicates inefficient budget usage, and reducing spend or optimizing this campaign would improve overall performance.
