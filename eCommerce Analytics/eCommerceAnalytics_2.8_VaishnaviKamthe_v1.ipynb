```python
journeys = [
    {
        "journey": ["Email", "Web", "Purchase"],
        "revenue": 100
    },
    {
        "journey": ["Social Media", "Email", "Web", "Purchase"],
        "revenue": 150
    },
    {
        "journey": ["Ads", "Web", "Purchase"],
        "revenue": 120
    }
]
```

The dataset does not contain explicit multi-touch customer journeys. Therefore, representative journeys were constructed using common marketing channels observed in the dataset to simulate realistic customer paths leading to conversion.

```python
def first_touch(journey, revenue):
    return {journey[0]: revenue}

def last_touch(journey, revenue):
    return {journey[-2]: revenue}  # last before "Purchase"

def linear_touch(journey, revenue):
    channels = journey[:-1]  # exclude "Purchase"
    share = revenue / len(channels)
    return {ch: share for ch in channels}

results = []

for j in journeys:
    journey = j["journey"]
    revenue = j["revenue"]
    
    results.append({
        "Journey": journey,
        "First Touch": first_touch(journey, revenue),
        "Last Touch": last_touch(journey, revenue),
        "Linear": linear_touch(journey, revenue)
    })

results
```

# Attribution Comparison

For the same customer journey, attribution varies significantly depending on the model used:

- In the first-touch model, the entire revenue is attributed to the first interaction (e.g., Email or Social Media), emphasizing acquisition channels.
- In the last-touch model, revenue is assigned to the final interaction before conversion (e.g., Web), highlighting closing channels.
- In the linear model, revenue is distributed equally across all touchpoints, providing a balanced view of channel contribution.

This comparison shows how attribution models can lead to different interpretations of channel effectiveness.

# Attribution Insight

In all journeys, the Web channel gets full credit in the last-touch model, even though earlier channels like Email or Social Media helped bring the customer.
This shows that last-touch overvalues closing channels and ignores discovery channels.
The first-touch model gives full credit to the starting channel, but ignores the role of later interactions in converting the customer.
The linear model distributes credit evenly, but it assumes all touchpoints are equally important, which may not be true in real scenarios.
